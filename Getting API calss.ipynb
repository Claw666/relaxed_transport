```python
def co2Emissions(mode, distance):
    estimateCO2 = estimate_co2(mode= mode, distance_in_km = distance)
    estimateCO2kg = estimateCO2/1000
    
    treesPlant = round(estimateCO2kg/20,2)
    
    rijksmuseumLightDays = round(estimateCO2kg/1.1,2)
    
    return estimateCO2kg, treesPlant, rijksmuseumLightDays
```

```python
def caloriesBurnend(sex, exercise, duration):
    averageHeightF = 169.3
    averageHeightM = 182.9
    averageWeightF = 72
    averageWeightM = 85
    
    if exercise == "walking":
        if sex == "female":
            caloriesBurnend = ((4.3*averageWeightF*3.5)/200)*duration
        else:
            caloriesBurnend = ((4.3*averageWeightM*3.5)/200)*duration
    
    elif exercise == "running":
        if sex == "female":
            caloriesBurnend = 11.5*3.5*(averageWeightF/200)*duration
        else:
            caloriesBurnend = 11.5*3.5*(averageWeightM/200)*duration

    elif exercise == "biking":
        if sex == "female":
            caloriesBurnend = (averageWeightF/2.20462)*11*0.0175*duration
        else:
            caloriesBurnend = (averageWeightM/2.20462)*11*0.0175*duration
                             
    return caloriesBurnend
```

```python
def placesGoogle(placeCategory, lat, long, radius):

    API_key = "AIzaSyD3DaloR43XmrjIYBYYcqh7zM1DGKzbdLA"
    category = placeCategory
    lat= str(lat)
    long = str(long)
    radius = str(radius)
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword=" + category + "&location=" +long +"%2C" + lat +"&radius=" + radius +"&type=" + category +"&key=" + API_key
    
    GoogleResponse = requests.get(url)
    json_GoogleObject = GoogleResponse.json()

    list_var=[]
    for result in np.arange(len(json_GoogleObject)):
        place = json_GoogleObject['results'][result]
        business_status = place['business_status']
        name = place['name']
        lat = place['geometry']['location']['lat']
        long = place['geometry']['location']['lng'] 
        userRating = place['rating']
        totalRating = place['user_ratings_total']
        if business_status == "OPERATIONAL":
            places_list = [name, lat, long, userRating, totalRating]
            list_var.append(places_list)
        else:
            pass
    return list_var
```
